```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('vgsales/vgsales.csv')
df.head(10)
```

## 1. Which company is the most common video game publisher?

```python

most_common_publisher = df['Publisher'].value_counts().idxmax()
most_common_publisher
```

## 2. Whatâ€™s the most common platform?

```python
most_common_platform = df['Platform'].value_counts().idxmax()
most_common_platform
```

## 3. What about the most common genre?

```python
most_common_genre = df['Genre'].value_counts().idxmax()
most_common_genre
```

## 4. What are the top 20 highest grossing games?

```python
top_twenty_highest_grossing_games = df.sort_values('Global_Sales', ascending=False).head(20)
top_twenty_highest_grossing_games
```

## 5. For North American video game sales, whatâ€™s the median?
* Provide a secondary output showing ten games surrounding the median sales output
* assume that games with same median value are sorted in descending order

```python
na_median = df['NA_Sales'].median()
print(na_median)

# games_around_median = df[df['NA_Sales'] == na_median]
# games_around_median.head(10)
df.iloc[(df['NA_Sales']- na_median).abs().argsort()[:11]]
# [midpoint_na_sales-5: midpoint_na_sales+5]


```

## 6. For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
top_game = df['Global_Sales'].max()

top_game_mean = df.nlargest(1, 'Global_Sales')
mean = top_game_mean.mean()['NA_Sales']

std = df['NA_Sales'].std()

z_scores = (top_game - mean) / std
print(z_scores)
```

## 7. The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
# wii_of_sales = wii_avg / (wii_avg + others_avg)

wii_sales  = df.loc[df['Platform'] == 'Wii']
wii_average = wii_sales['Global_Sales'].mean()


not_wii_sales = df.loc[df['Platform'] != 'Wii']
not_wii_average = not_wii_sales['Global_Sales'].mean()

wii_of_sales = wii_average / (wii_average + not_wii_average)
wii_of_sales
```
